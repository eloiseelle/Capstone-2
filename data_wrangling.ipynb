```python
# Standardize city names
def clean_city(city):
    if pd.isnull(city):
        return None
    return city.split(',')[0].strip().lower()

tidal_df['city_clean'] = tidal_df['city'].apply(clean_city)
spotify_hiphop_df['city_clean'] = spotify_hiphop_df['city'].apply(clean_city)
spotify_10k_df['city_clean'] = spotify_10k_df['city'].apply(clean_city)
complex_cities_df['city_clean'] = complex_cities_df['City'].apply(clean_city)
billboard_df['city_clean'] = billboard_df['city'].apply(clean_city)
```

```python
# Clean artist and city
merged_df['artist'] = merged_df['artist'].str.strip().str.lower()
merged_df['city_clean'] = merged_df['city_clean'].str.strip().str.lower()

# Combine all NYC boroughs
def merge_nyc(city):
    if city in ['queens', 'new york', 'brooklyn', 'bronx', 'harlem', 'manhattan']:
        return 'nyc'
    return city

# Create a city column 'NYC' 
merged_df['city_grouped'] = merged_df['city_clean'].apply(merge_nyc)

# Group by artist and city_grouped
grouped_artists_by_city = merged_df.groupby(['city_grouped', 'artist']).sum(numeric_only=True).reset_index()

# Recalculate total 
grouped_artists_by_city['total_count'] = grouped_artists_by_city[
    ['tidal_count', 'hiphop_count', 'top10k_count', 'complex_count', 'billboard_count']
].sum(axis=1)

# Sort results to get top female artists by total count
top_by_city = grouped_artists_by_city.sort_values(by='total_count', ascending=False)
top_by_city.head(10)
```
